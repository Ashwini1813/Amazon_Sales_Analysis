```python
df = df.dropna()
```

```python
df = df.drop(['img_link','product_link','review_title','review_content','about_product'], axis = 1)
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)
df['rating_count'] = df['rating_count'].str.replace(',','').astype(float)
df['rating'] = pd.to_numeric(df['rating'], errors = 'coerce')
```

```python
df['main_category'] = df['category'].str.split('|').str[0]
print(df['main_category'].value_counts())
```

Now we use SQL 

```python
import sqlite3
```

```python
conn = sqlite3.connect('amazon_sales_dataset.db')
```

```python
df.to_sql('amazon_sales_dataset', conn, if_exists = 'replace', index = False)
```

```python
print("Total rows: ", pd.read_sql("SELECT COUNT(*) FROM amazon_sales_dataset", conn).iloc[0,0])
```

```python
result_check = pd.read_sql("""
    SELECT * FROM amazon_sales_dataset LIMIT 2 """, conn)
print(result_check.columns.tolist())
```

Query 1: Top 10 Highest Rated Products

```python
result1 = pd.read_sql("""
    SELECT product_name, rating, rating_count
    FROM amazon_sales_dataset
    ORDER BY rating DESC
    LIMIT 10
    """, conn)
print(result1)
```

Query 2: Average Rating According to Category 

```python
result2 = pd.read_sql("""
    SELECT main_category,
            COUNT(*) AS total_products,
            ROUND(AVG(rating),2) AS avg_rating,
            ROUND(AVG(discount_percentage),2) AS avg_discount
    FROM amazon_sales_dataset
    GROUP BY main_category
    ORDER BY avg_rating DESC
    LIMIT 10
    """, conn)
print(result2)
```

Query 3: Top 10 Most Discounted Products

```python
result3 = pd.read_sql("""
    SELECT product_name, actual_price, discounted_price, discount_percentage, rating
    FROM amazon_sales_dataset
    WHERE discount_percentage > 50
    ORDER BY discount_percentage DESC
    LIMIT 10
    """, conn)
print(result3)
```

Query 4: Price Range Analysis

```python
result4 = pd.read_sql("""
    SELECT
        CASE
            WHEN discounted_price < 500 THEN 'Budget (< 500)'
            WHEN discounted_price BETWEEN 500 AND 2000 THEN 'Mid Range (500-2000)'
            WHEN discounted_price BETWEEN 2000 AND 10000 THEN 'Premium (2000-10000)'
            ELSE 'Luxury (> 10000)'
        END as price_range,
        COUNT(*) AS total_products,
        ROUND(AVG(rating), 2) as avg_rating
    FROM amazon_sales_dataset
    GROUP BY price_range
    ORDER BY total_products DESC
    """,conn)
print(result4)
```

Query 5: Discount VS Rating Correlation

```python
result5 = pd.read_sql("""
    SELECT
        CASE
            WHEN discount_percentage < 20 THEN 'Low Discount (< 20%)'
            WHEN discount_percentage BETWEEN 20 AND 50 THEN 'Medium Discount (20-50%)'
            ELSE 'High Discount (> 50%)'
        END AS discount_range,
        COUNT(*) AS total_products,
        ROUND(AVG(rating), 2) AS avg_rating,
        ROUND(AVG(discounted_price), 2) AS avg_price
    FROM amazon_sales_dataset
    GROUP BY discount_range
    ORDER BY avg_rating DESC
    """, conn)
print(result5)
```
